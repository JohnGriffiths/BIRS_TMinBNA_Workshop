````python
fig, ax = plt.subplots(ncols=5,figsize=(12,3))

for f_it,(k,v) in enumerate(meg_blpcys.items()):
    a = ax.ravel()[f_it]
    img = nib.gifti.giftiio.read(v)
    dat = np.squeeze(img.get_data())
    a.imshow(dat,cmap='coolwarm')
    a.set_title(k)
    a.axis('off')
````
